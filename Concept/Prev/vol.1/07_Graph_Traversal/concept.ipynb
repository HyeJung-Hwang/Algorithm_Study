```python
# bfs
# 방문할 자식 노드 저장하기 위한 자료구조 => 큐
# 첫 시작이 특정 노드일 수 있음
from collections import deque                                                                        
                                                                                                    
def bfs(graph, start):                                                                               
    visited = set([start])                                                                           
    queue = deque([start])

    while queue:
        node = queue.popleft()
        print(node)  # 처리

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor) # 방문 경로에 들어갔었는지 를 검사
                queue.append(neighbor) 
```

```python
# DFS
# 탐색 대상 = 그래프 (문제에서 주어지는 입력)
# 그래프 방문 경로 = 리스트
def dfs(graph, start):
    visited = set([start]) # 방문 여부 저장, 해시로 바로 찾음
    stack = [start] # 임시 자료구조

    while stack:
        node = stack.pop()
        print(node)  # 처리
        # 탐색 파트, 가장 최근꺼에 대해서 계속 싶게 , pop해서, 주위 이웃들 스택에 넣기  또 pop해서 주위이웃들 스택에 넣기 
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append(neighbor)
```
