```python
import queue
```

```python
normal_queue = queue.Queue()
for i in range(5):
    normal_queue.put(i)
while normal_queue.qsize(): # [Mistake] 비어있는지 참거짓으로 while 돌릴 떄 길이 활용하기 
    print(normal_queue.get())

if normal_queue.empty():
    print("Queue is empty")
```

```python
normal_queue = queue.Queue()
for i in range(5):
    normal_queue.put(i)
while  not normal_queue.empty(): # [Mistake] 비어있는지 참거짓으로 while 돌릴 떄 비어있는 지 여부 자체를  활용하기 
    print(normal_queue.get())

if normal_queue.empty():
    print("Queue is empty")
```
